```python
# Import libraries
import os
import numpy as np
import pandas as pd
import tensorflow as tf
import random
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import layers
from tensorflow.keras.layers import LSTM, Dense, Dropout
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
import requests
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
from tabulate import tabulate
import tensorflowjs as tfjs
from alpaca_trade_api import REST
from alpaca.data.historical import CryptoHistoricalDataClient
from alpaca.data.requests import CryptoBarsRequest, StockLatestQuoteRequest
from alpaca.data.timeframe import TimeFrame
from alpaca.data.live import StockDataStream
from alpaca.trading.client import TradingClient
from alpaca.trading.requests import MarketOrderRequest
from alpaca.trading.enums import OrderSide, TimeInForce
from alpaca.data.historical import StockHistoricalDataClient
from alpaca.trading.requests import LimitOrderRequest
from alpaca.trading.requests import GetOrdersRequest
from alpaca.trading.enums import OrderSide, QueryOrderStatus
import math



ALPACA_API_KEY = "PKWZQPWFN48ANDCSHDJZ"
ALPACA_API_SECRET = "a88iGvrTfywVtrcQxO0vnsE9xwlSKtIjkRpcIf6C"
ALPACA_BASE_URL = "https://paper-api.alpaca.markets"

ALPACA_CREDS = {
    "API_KEY":ALPACA_API_KEY,
    "API_SECRET": ALPACA_API_SECRET,
    "PAPER": True
}

api = REST(base_url=ALPACA_BASE_URL, key_id=ALPACA_API_KEY, secret_key=ALPACA_API_SECRET)
trading_client = TradingClient(ALPACA_API_KEY, ALPACA_API_SECRET, paper=True)
client_hist = StockHistoricalDataClient(ALPACA_API_KEY, ALPACA_API_SECRET)


#FMP API for training data
api_key_fmp = "dc16ae239a0a90cc7f039177aa18aa33"

### TRAINING CODE

def fetch_stock_data(symbol, from_date, to_date):
    api_url = f"https://financialmodelingprep.com/api/v3/historical-price-full/{symbol}?apikey={api_key_fmp}&from={from_date}&to={to_date}"
    response = requests.get(api_url)
    data = response.json()
    if data:
        historical_data = data['historical']

        reversed_data = historical_data[::-1]
        return reversed_data
    else:
      print(f"Error in data fetching: {data}")
      return data

# Define function for data preprocessing
def preprocess_data(data):
    closing_prices = [day['close'] for day in data]
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(np.array(closing_prices).reshape(-1, 1))
    return scaled_data, scaler, [day['date'] for day in data], closing_prices

# Define function to create sequences for LSTM
def create_sequences(data, seq_length):
    X, y = [], []
    # for i in range(len(data) - seq_length):
    #     X.append(data[i:(i + seq_length)])
    #     y.append(data[i + seq_length - 1])
    # return np.array(X), np.array(y)
     # Pad the data with zeros at the beginning
    padded_data = np.pad(data, ((seq_length - 1, 0), (0, 0)), mode='constant', constant_values=0.5)

    # Iterate through the padded data to create sequences
    for i in range(len(data)):
        sequence = padded_data[i:(i + seq_length)]
        X.append(sequence)
        y.append(data[i])

    return np.array(X), np.array(y)

# Define function to build LSTM model
def build_lstm_model(seq_length):

    model = Sequential([layers.Input((seq_length, 1)),
                    layers.LSTM(64),
                    layers.Dense(32, activation='relu'),
                    layers.Dense(32, activation='relu'),
                    layers.Dropout(0.2),
                    layers.Dense(1)])

    model.compile(loss='mse',
              optimizer=Adam(learning_rate=0.001),
              metrics=['mean_absolute_error'])
    return model

def build_redundant_lstm_models(seq_length, num_models):
    models = []
    for _ in range(num_models):
        model = build_lstm_model(seq_length)
        models.append(model)
    return models

# Define function for training the model
def train_model(model, X_train, y_train, X_val, y_val, epochs, batch_size):
    model.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=epochs, batch_size=batch_size, verbose=0)
    return model

# Define function to save the trained model
def save_model(model, filename):
    model.save(filename + '.h5')  # Save the model in the native Keras format


# Define function to calculate model accuracy
def calculate_accuracy(model, X_test, y_test):
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    return rmse

def run_code(symbol, start_date, end_date, future_days):
    seq_length = 20
    epochs = 100
    batch_size = 32
    model_filename = 'stock_prediction_model'

    # Fetch data
    stock_data = fetch_stock_data(symbol, start_date, end_date)

    # Preprocess data
    scaled_data, scaler, dates, closing_prices = preprocess_data(stock_data)

    # Create sequences
    X, y = create_sequences(scaled_data, seq_length)

    # Split data into training, validation, and testing sets
    q_80 = int(len(X) * 0.8)
    q_90 = int(len(X) * 0.9)


    X_train, y_train = X[:q_80], y[:q_80]
    X_val, y_val = X[q_80:q_90], y[q_80:q_90]
    X_test, y_test = X[q_90:], y[q_90:]


    dates_train = [dates[i] for i in range(q_80)]
    dates_val = [dates[q_80 + i] for i in range(len(y_val))]
    dates_test = [dates[q_90 + i] for i in range(len(y_test))]



    dates_train = [datetime.strptime(date_str, '%Y-%m-%d').date() for date_str in dates_train]
    dates_val = [datetime.strptime(date_str, '%Y-%m-%d').date() for date_str in dates_val]
    dates_test = [datetime.strptime(date_str, '%Y-%m-%d').date() for date_str in dates_test]

    # # Plotting
    # plt.plot(dates_train, y_train, color='orange')
    # plt.plot(dates_val, y_val, color='red')
    # plt.plot(dates_test, y_test, color='green')
    # plt.title(f'{symbol} Data')

    # plt.legend(['Train', 'Validation', 'Test'])
    # plt.show()

    # Reshape data for LSTM
    X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
    X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))

    # Build LSTM model
    model = build_lstm_model(seq_length)

    # Train model
    model = train_model(model, X_train, y_train, X_val, y_val, epochs, batch_size)

    rmse = calculate_accuracy(model, X_test, y_test)

    print("accuracy", rmse)

    # Save model
    # save_model(model, model_filename)
    # tf.saved_model.save(model, './')
    tfjs.converters.save_keras_model(model, "./model_tfjs")

    # Predict using the trained model
    y_train_pred = model.predict(X_train)
    y_val_pred = model.predict(X_val)
    y_test_pred = model.predict(X_test)


    # Inverse transform the predicted and actual values
    y_train_actual = scaler.inverse_transform(y_train.reshape(-1, 1))
    y_train_pred_actual = scaler.inverse_transform(y_train_pred)

    y_val_actual = scaler.inverse_transform(y_val.reshape(-1, 1))
    y_val_pred_actual = scaler.inverse_transform(y_val_pred)

    y_test_actual = scaler.inverse_transform(y_test.reshape(-1, 1))
    y_test_pred_actual = scaler.inverse_transform(y_test_pred)


    plt.figure(figsize=(10, 6))

    # Plot training data
    plt.plot(dates_train, y_train_actual,color='blue', label='Training Actual')
    plt.plot(dates_train, y_train_pred_actual, color='orange', label='Training Predicted')

    # Plot validation data
    plt.plot(dates_val, y_val_actual, color='blue', label='Validation Actual')
    plt.plot(dates_val, y_val_pred_actual, color='red', label='Validation Predicted')

    # Plot testing data
    plt.plot(dates_test, y_test_actual,color='blue', label='Testing Actual')
    plt.plot(dates_test, y_test_pred_actual, color='green', label='Testing Predicted')


    # Extend predictions for additional days
    future_prices = []

    X_extend = X[-3:].copy()
    inverse_X = scaler.inverse_transform(X.reshape(-1, 1))

    for _ in range(future_days):  # Predict for the next 7 days
        future_price = model.predict(X_extend)[0][0]

        future_prices.append(future_price)
        X_extend = np.roll(X_extend, -1)
        X_extend[-1][-1] = future_price
        # print(X_extend)

    future_prices_actual = scaler.inverse_transform(np.array(future_prices).reshape(-1, 1))

    # # Extend predictions
    future_dates = generate_date_range(dates_test[-1] + timedelta(days=1), dates_test[-1] + timedelta(days=future_days))

    plt.plot(future_dates, future_prices_actual, color='purple', label='Future Predictions')

    plt.title(f'{symbol} Actual vs. Predicted Closing Prices')
    plt.xlabel('Date')
    plt.ylabel('Closing Price $')
    plt.xticks(rotation=45)
    plt.legend()
    plt.tight_layout()
    plt.show()

    return dates_test, closing_prices, future_prices_actual, rmse


def generate_date_range(start_date, end_date):
    return [start_date + timedelta(days=i) for i in range((end_date - start_date).days + 1)]

########## Stock API Functions ##############

def get_portfolio_symbols():
    # Get list of positions
    positions = api.list_positions()

    return positions

def execute_trade(symbol, quantity, side):
    market_order_data = MarketOrderRequest(
                            symbol=symbol,
                            qty=quantity,
                            side=side,
                            time_in_force=TimeInForce.DAY
                        )
    market_order = trading_client.submit_order(order_data=market_order_data)


def main():

    today_date = datetime.today()
    six_months_ago = today_date - timedelta(days=30*12)
    # Format today's date as 'YYYY-MM-DD'
    start_date = six_months_ago.strftime('%Y-%m-%d')
    end_date = today_date.strftime('%Y-%m-%d')
    future_days = 3

    #Get previous order prices
    portfolio = get_portfolio_symbols()
    request_params = GetOrdersRequest(
                    status=QueryOrderStatus.CLOSED,
                    side=OrderSide.BUY
        )
    orders = trading_client.get_orders(filter=request_params)

    previous_orders = {}
    current_prices = {}

    # Iterate over each order dictionary
    for order in orders:
        symbol = order.symbol
        filled_avg_price = float(order.filled_avg_price)
        filled_qty = int(order.filled_qty)
        # Store the previous orders in the dictionary
        previous_orders[symbol] = {'filled_avg_price': filled_avg_price, 'filled_qty': filled_qty}

        multisymbol_request_params = StockLatestQuoteRequest(symbol_or_symbols=symbol)
        latest_multisymbol_quotes = client_hist.get_stock_latest_quote(multisymbol_request_params)

        current_prices[symbol] = {'ask_price': latest_multisymbol_quotes[symbol].ask_price}


    for position in portfolio:
        symbol = position.symbol
        quantity = position.qty

        print(symbol)

        # relevant prices for the stock

        filled_avg_price = float(previous_orders[symbol]['filled_avg_price'])
        filled_qty = int(previous_orders[symbol]['filled_qty'])
        curr_price = float(current_prices[symbol]['ask_price'])


        print("price bought at:", filled_avg_price)

        print("current stock price:" , curr_price)

        # Calculate stop-loss and take-profit prices
        stop_loss_price = filled_avg_price * 0.90 # 5% below filled average price
        take_profit_price = filled_avg_price * 1.10  # 5% above filled average price

        print("stop_loss", stop_loss_price)
        print("takeprofit", take_profit_price)

        # # Current qty and stock trades based on stop loss and profit
        if (curr_price < stop_loss_price ):
          print("Cut our losses")
          execute_trade(symbol, quantity, 'sell')


        if (curr_price > take_profit_price ):
          print("Take our profits")
          execute_trade(symbol, quantity, 'sell')


        # Adjust quantity based on symbol weight
        quantity_trade_with = math.floor(int(filled_qty/2))

        profit_curr_trade = math.floor(int(curr_price * 1.10))
        stop_loss_curr_trade = math.floor(int(curr_price * 0.9))


        # Prediction Code

        dates_test, closing_prices, future_prices_actual, accuracy = run_code(symbol, start_date, end_date, future_days)
        print(closing_prices)
        print(future_prices_actual)

        # Determine future trend is upwards buy and also check accuracy is above 90%
        if future_prices_actual[-1] > future_prices_actual[0] and accuracy < 0.1:
            # Buy order with stop loss and take profit
            api.submit_order(
                symbol=symbol,
                side='buy',
                type='market',
                qty=quantity_trade_with,
                time_in_force='day',
                order_class='bracket',
                take_profit=dict(
                    limit_price=str(profit_curr_trade),  # Example: Take profit at 10% above average filled price
                ),
                stop_loss=dict(
                    stop_price=str(stop_loss_curr_trade),  # Example: Stop loss at 10% below average filled price
                    limit_price=str(stop_loss_curr_trade),  # Same as stop_price for market orders
                )
            )
        #if trend is downwards and accuracy is above 90%
        elif future_prices_actual[0] > future_prices_actual[-1] and accuracy < 0.1:
            #sell buy not all since its not our stop_loss but just a prediction
            execute_trade(symbol, quantity_trade_with, 'sell')

if __name__ == "__main__":
    main()
```
